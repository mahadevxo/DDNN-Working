```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_inf_time(p, device_perf):
    p*=100
    slope = 0.0003872
    y_intercept = -0.0003872 
    t_nominal = slope * p + y_intercept
    t_scaled = t_nominal + ((1-device_perf))
    t_scaled = np.maximum(t_scaled, 0)  # Ensure no negative times

    return t_scaled*1000
```

```python
plt.plot(np.arange(0, 1.01, 0.01), get_inf_time(np.arange(0, 1.01, 0.01), 0.5), label='Device Perf = 0.5')
```
