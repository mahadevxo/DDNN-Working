```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vals = np.random.beta(2, 8, int(1e5))
plt.figure(figsize=(10, 6))
plt.hist(vals, bins=1000, density=True)
plt.title("Histogram of Beta Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
A_MIN = 50
SIGMA_R=4
SIGMA_L=12

def R_A(A_p):
    delta_a = A_p - A_MIN
    if 0<= delta_a <= 1e-2:
        return np.exp(- (delta_a**2) / (10 * SIGMA_L**2)) * 100
    elif delta_a > 1e-2:
        return np.exp(- (delta_a) / (2 * SIGMA_L**2)) * 100
    else:
        return np.exp(- (abs(delta_a)**1.5) / (2 * SIGMA_R**2)) * 100

A_p = np.linspace(1, 0, num=1000)**3 * 100
R_A_vals = [R_A(a) for a in A_p]
# len(R_A_vals), len(A_p)

plt.figure(figsize=(10, 6))
plt.plot(A_p, R_A_vals, label=f'Accuracy Reward Function\nA_MIN=60, SIGMA_R={SIGMA_R}, SIGMA_L={SIGMA_L}')
plt.title("R_A Function")
plt.xlabel("A_p")
plt.ylabel("R_A Value")
plt.grid()
plt.legend()
plt.show()
```
