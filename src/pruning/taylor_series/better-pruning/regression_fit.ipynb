```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('results.csv')
```

```python
data
```

```python
#make 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
# Plot the first subplot
axs[0].plot(data['pruning_amount'], data['accuracy'], label='Accuracy', color='blue')
axs[0].set_title('Accuracy vs Pruning Amount')
axs[0].set_xlabel('Pruning Amount')
axs[0].set_ylabel('Accuracy')
# Plot the second subplot
axs[1].plot(data['pruning_amount'], data['model_size'], label='Model Size', color='red')
axs[1].set_title('Model Size vs Pruning Amount')
axs[1].set_xlabel('Pruning Amount')
axs[1].set_ylabel('Model Size (MB)')
# Plot the third subplot
axs[2].plot(data['pruning_amount'], data['comp_time'], label='Inference Time', color='green')
axs[2].set_title('Inference Time vs Pruning Amount')
axs[2].set_xlabel('Pruning Amount')
axs[2].set_ylabel('Inference Time (ms)')
# Add legends to each subplot
axs[0].legend()
axs[1].legend()
axs[2].legend()
# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

```python
def polynomial_fit(x, y, degree=2):
    """Fit a polynomial of given degree to the data."""
    coeffs = np.polyfit(x, y, degree)
    return np.poly1d(coeffs)
```

```python
polyfit_accuracy = polynomial_fit(data['pruning_amount'], data['accuracy'], degree=5)
polyfit_accuracy
```

```python
plt.plot(data['pruning_amount'], data['accuracy'], label='Data Points')
plt.plot(data['pruning_amount'], polyfit_accuracy(data['pruning_amount']), label='Polynomial Fit', color='orange')
plt.title('Polynomial Fit for Accuracy vs Pruning Amount')
plt.xlabel('Pruning Amount')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
